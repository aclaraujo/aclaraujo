## Exibindo dados com matplotlib

A bilioteca matplotlib permite a construção de gráficos a partir dos dados de um _DataFrame_. Para isso, basta importar a biblioteca da forma:

``` from matplotlib import pyplot as plt ```

````python
from matplotlib import pyplot as plt
````

````python
plt.figure(figsize=(10,8))
#plt.scatter('idade_conc','qt_sm', data=df[['idade_conc','qt_sm']])
plt.scatter(df['idade_conc'], df['qt_sm'])
````

> O objeto ` DataFrame ` possui um wrapper para o pacote ` pyplot `, permitindo que os gráicos sejam criados a partir dele

````python
df[['especie','qt_sm']].groupby('especie', as_index=False).mean().plot.barh(x='especie',y='qt_sm', figsize=(10,10))
````

> Subdivisões nos eixos

````python
dados = df[['especie','sexo','idade_conc']].groupby(['especie','sexo']).idade_conc.mean()
dados.sort_values().nlargest(10).unstack().plot.barh(stacked=False, figsize=(10,10))
````

> Duas séries em um mesmo gráfico

````python
df_aux = df[df.especie == 'Auxílio Doenca Previdenciário']

dt1 = df[['sigla_uf','valor_rend']].groupby(['sigla_uf']).sum()
dt2 = df_aux[['sigla_uf','valor_rend']].groupby(['sigla_uf']).sum()

fig, axs = plt.subplots(figsize=(10,10))

axs.bar(dt1.index,dt1['valor_rend'], label='Total')
axs.plot(dt2.index,dt2['valor_rend'], label='Auxílio Doenca Previdenciário', c='r')

plt.legend()
````

# Analisando dados da Covid-19

## Comparando países

````python
from datetime import date, datetime, timedelta
import numpy as np
from matplotlib.ticker import ScalarFormatter
from scipy.ndimage.filters import gaussian_filter1d
from math import ceil
import requests
````

````python
df = pd.read_csv('https://covid19.who.int/WHO-COVID-19-global-data.csv', 
                 skipinitialspace=True, dtype={'Country':'string','Country_code':'string', 'WHO_region': 'string'},  
                 parse_dates=['Date_reported'])
df_covid = df[df['Cumulative_cases'] >= 100]
````

````python
df_pop = pd.read_csv("/content/drive/MyDrive/who_population.csv", thousands=' ', dtype={'Population (in thousands) total':'float'})
df_pop = df_pop.groupby('Country', as_index=False).first()
df_pop['Population'] = df_pop['Population (in thousands) total'].astype('int')*1000
````

````python
df_codiv_data = df_covid.merge(df_pop, on='Country', how='left')
df_codiv_data.set_index(['Country','Date_reported'], inplace=True)
df_codiv_data[['Inc_Cumulative_cases', 'Inc_New_cases','Inc_Cumulative_deaths','Inc_New_deaths']] = df_codiv_data[['Cumulative_cases', 'New_cases','Cumulative_deaths','New_deaths']].apply(lambda x: x*100000/df_codiv_data['Population'])
df_codiv_data
````

````python
df_codiv_data_10_larg = df_codiv_data.loc[df_codiv_data.groupby('Country').max().nlargest(10, 'Cumulative_deaths').index]
df_codiv_data_10_larg = df_codiv_data_10_larg[['Inc_New_cases','Inc_New_deaths','Cumulative_deaths']].stack().unstack(level=2)
````

````python
countries = df_codiv_data_10_larg.index.levels[0]
#countries = ['Brazil','Spain','United States of America','France','The United Kingdom']

_, axs = plt.subplots(3,1, figsize=(15,15))

for country in countries:
    
    data = df_codiv_data_10_larg.loc[country]
    
    data.reset_index(inplace=True)
    data.sort_values('Date_reported',inplace=True)
    data.index = pd.to_timedelta(data['Date_reported'].apply(lambda x: x - data.Date_reported.min()))
    x = data.index.days
    
    y = gaussian_filter1d(data['Inc_New_cases'], sigma=3)
    axs[0].plot(x, y, label=country)
    
    y = gaussian_filter1d(data['Inc_New_deaths'], sigma=3)
    axs[1].plot(x, y, label=country)
    
    y = gaussian_filter1d(data['Cumulative_deaths'], sigma=3)
    axs[2].plot(x, y, label=country)
    axs[2].set_yscale('log')
    #plt.sca(axs[2])
    axs[2].yaxis.set_major_formatter(ScalarFormatter())
    
plt.legend()
````
